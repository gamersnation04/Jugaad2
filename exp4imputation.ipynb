```python
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('MEDAL1DATA.csv')
df.head()
```

```python
#Fill with default value
df2 = df.fillna(value = 0)
df2.head()
```

```python
#Fill with previous value
df2 = df.fillna(method = 'pad')
df2.head()
```

```python
#Fill with next value
df3 = df.fillna(method = 'bfill')
df3.head()
```

```python
#Fill with mean
def get_dict(dataFrame, ignoreList, strategy):
    dictionary = {}
    for series_name, series in dataFrame.items():
        if series_name in ignoreList:
            continue
        if strategy == "mean":
            ans = dataFrame[series_name].mean()
        elif strategy == "median":
            ans = dataFrame[series_name].median()
        elif strategy == "mode":
            ans = dataFrame[series_name].mode().max()
        elif strategy == "max":
            ans = dataFrame[series_name].max()
        elif strategy == "min":
            ans = dataFrame[series_name].min()
        dictionary[series_name] = ans
    return dictionary
    
df4 = df.fillna(get_dict(df, ["STATE"], "mean"))
df4.head()
```

```python
#Fill with median
df5 = df.fillna(get_dict(df, ["STATE"], "median"))
df5.head()
```

```python
#Fill with Mode
df6 = df.fillna(get_dict(df, ["STATE"], "mode"))
df6.head()
```

```python
#Fill with Max
df7 = df.fillna(get_dict(df, ["STATE"], "max"))
df7.head()
```

```python
#Fill with min
df7 = df.fillna(get_dict(df, ["STATE"], "min"))
df7.head()
```

```python
#Fill with Most Frequent
from sklearn.impute import SimpleImputer
import pandas as pd

df9 = df.drop("STATE", axis="columns")
imp1 = SimpleImputer(strategy="most_frequent")
res1 = imp1.fit_transform(df9)
newData1 = pd.DataFrame(res1)
newData1.head()
```

```python
#Fill using KNN
df10 = df.drop("STATE", axis="columns")
imp2 = KNNImputer(n_neighbors=2)
res2 = imp2.fit_transform(df10)
newData2 = pd.DataFrame(res2)
newData2.tail()
```

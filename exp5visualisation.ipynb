```python
# Calculating total scores of each student
df['TOTAL'] = df[['ADS', 'SMA', 'DC', 'EM']].sum(axis=1)

# Calculating pass/fail count
pass_count = df[df['TOTAL'] > 134].shape[0]
fail_count = df[df['TOTAL'] <= 134].shape[0]
```

```python
# Plotting Total Scores of Students
plt.figure(figsize=(12, 6))
plt.bar(df.index, df['TOTAL'], color='skyblue')
plt.xlabel('Students')
plt.ylabel('Total Score')
plt.title('Total Scores of Students')
plt.xticks(df.index, df['NAME '], rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv("MARKS2.csv")

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(df['TOTAL'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('TOTAL Marks')
plt.ylabel('Frequency')
plt.title('Histogram of TOTAL Marks')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sorting the dataframe by total marks
df_sorted = df.sort_values(by='TOTAL', ascending=False)

# Plotting the stacked area chart
plt.figure(figsize=(12, 6))
plt.stackplot(df_sorted['NAME '], df_sorted['TOTAL'], labels=['Total Marks'])
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.title('Stacked Area Chart of Total Marks for Students')
plt.legend(loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv("MARKS2.csv")

# Extract ADS marks and total marks
ads_marks = df['ADS']
total_marks = df['TOTAL']

# Plotting the box plot
plt.figure(figsize=(8, 6))
plt.boxplot([ads_marks, total_marks], labels=['ADS Marks', 'Total Marks'])
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Box Plot of ADS Marks and Total Marks')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Selecting the columns for the violin plot
cols = ['ADS', 'SMA', 'EM', 'DC', 'TOTAL']

# Melting the dataframe for plotting
df_melted = df.melt(value_vars=cols, var_name='Subject', value_name='Marks')

# Plotting the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Subject', y='Marks', data=df_melted, palette='muted')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Violin Plot of Marks for ADS, SMA, EM, DC, and Total')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv("MARKS2.csv")

# Sum up the marks for each subject
subject_sums = df[['ADS', 'SMA', 'EM', 'DC']].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(subject_sums, labels=subject_sums.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Marks Across All Subjects')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv("MARKS2.csv")

# Subset the dataframe to include only the required columns
df_heatmap = df[['ADS', 'SMA', 'EM', 'DC']]

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_heatmap.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Marks Distribution for ADS, SMA, EM, and DC')
plt.tight_layout()
plt.show()

```
